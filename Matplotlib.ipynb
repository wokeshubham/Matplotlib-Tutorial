```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

**Basic Graph**

```python
a = np.array([0,1,2,3,4])
b = np.array([0,5,6,7,8])

#Resize the graph
plt.figure(figsize=(5,3), dpi=100)

#keyword argument notation
# plt.plot(a,b, label = '2x', c = 'magenta', linewidth = 5, linestyle = '--', marker= '.', markersize = '20', markeredgecolor= 'black')

#Use shorthand marker
# '[color][marker][line]'
plt.plot(a,b, 'r^-.', label = '2x')

#Line 2

#select intervals we want points at
x2 = np.arange(0,4.5,0.5)

#plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'g-', label= 'x^2')

#plot part of the graph as dashed line
plt.plot(x2[5:], x2[5:]**2, 'g.--')

#Add Title(specify font parameters with fontdict)
plt.title("My first graph", fontdict={'fontname': 'Comic Sans MS', 'fontsize': '20', 'color': '#800080'})

#X and Y labels
plt.xlabel("X Axis", fontdict={'fontname': 'Comic Sans MS', 'fontsize': '15'})
plt.ylabel("Y Axis", fontdict={'fontname': 'Comic Sans MS', 'fontsize': '15'})

#X and Y axis Tickmarks (scale of Graph)
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])

#Add a legend
plt.legend()
#plt.legend(loc='upper left', bbox_to_anchor=(1, 0.9), fancybox=True, shadow =True)

#Save Figure
plt.savefig('myfirstgraph.png', dpi=300)

#Show plot
plt.show()
```

**Barchart**

```python
label = ['A', 'B', 'C']
value = [1,4,2]

bars = plt.bar(label, value)

patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

#bars[0].set_hatch('/')
#bars[1].set_hatch('O')
#bars[2].set_hatch('*')

plt.figure(figsize=(5,3), dpi=200)

plt.show()
```

**Fifa Data**

```python
fifa = pd.read_csv('fifa_data.csv')

fifa.head(5)
```

**Histogram**

```python
plt.title('Player Ratings', fontdict={'fontsize': 20})

bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins= bins, color='r')

plt.xticks(bins)

plt.ylabel('Number of Players', fontdict={'fontsize': 15})
plt.xlabel('Overall Rating', fontdict={'fontsize': 15})

plt.show()
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

Labels = ['Left', 'Right']

colors = [ 'yellow','orange']

plt.pie([left,right], labels = Labels, colors= colors, radius=1, autopct= '%.2f')

plt.title("Preferred Legs of players")

plt.legend(loc = 'upper right', bbox_to_anchor= (1.3,1))

plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str  else x for x in fifa.Weight]

label = ['Light', 'Medium', 'Heavy']
color = ['#f54640', '#4287f5', '#46f540']

light = fifa.loc[fifa.Weight < 150].count()[0]
medium = fifa.loc[(fifa.Weight >=150) & (fifa.Weight <200)].count()[0]
heavy = fifa.loc[fifa.Weight >= 200].count()[0]

explode = [.3,0,.3]

plt.title("Weight Distribution of Fifa Players", loc='center')

plt.pie([light,medium,heavy], labels= label, colors=color, autopct= '%.2f', radius=1, explode = explode )

plt.legend(loc = 'upper right', bbox_to_anchor = (2,1))

plt.show()
```

```python
plt.figure(figsize=(5,6))

team1 = fifa.loc[fifa.Club == 'FC Barcelona' ]['Overall']
team2 = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
team3 = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

Labels = ['FC Barcelona', 'Real Madrid', 'NE Revolution']

boxes = plt.boxplot([team1,team2, team3], labels= Labels, patch_artist=True, medianprops={'linewidth':2})

for box in boxes['boxes']:
    box.set(color='red', linewidth=2)
    box.set(facecolor = '#e0e0e0')

plt.title('Professional Soccer Team Comparision')
plt.ylabel('FIFA Overall Rating')

plt.show()
```
